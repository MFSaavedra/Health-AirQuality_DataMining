## 1. Introducción

La contaminación atmosférica constituye uno de los principales problemas de salud pública en zonas urbanas. En la Región Metropolitana de Chile, los episodios de mala calidad del aire son frecuentes, especialmente durante el invierno, y se han asociado históricamente con un aumento en enfermedades respiratorias, hospitalizaciones y mortalidad.

Este proyecto busca analizar, desde una perspectiva de minería de datos, la relación entre distintos contaminantes atmosféricos y eventos de salud, utilizando tanto técnicas exploratorias como modelos estadísticos y de aprendizaje automático.

A diferencia de enfoques tradicionales centrados únicamente en correlaciones simples, este trabajo incorpora análisis temporal, espacial y metodologías supervisadas y no supervisadas para obtener una visión integral del fenómeno.

---

## 2. Objetivos

### 2.1 Objetivo general

Analizar la relación entre la calidad del aire y los eventos de salud respiratoria en la Región Metropolitana, mediante técnicas de minería de datos, modelamiento estadístico y aprendizaje automático.

### 2.2 Preguntas de investigación

1. **Asociación contaminantes–urgencias:** ¿Existe una asociación significativa entre los niveles diarios de contaminantes atmosféricos (PM2.5, PM10, NO₂, O₃, SO₂ y CO) y las atenciones de urgencia por enfermedades respiratorias en la Región Metropolitana?

2. **Impacto relativo y no linealidades:** ¿Qué contaminantes presentan el mayor impacto relativo sobre las defunciones y atenciones respiratorias, considerando efectos no lineales y posibles interacciones entre contaminantes?

3. **Efectos rezagados:** ¿Existen efectos rezagados entre la exposición a contaminación atmosférica y los eventos de salud respiratoria, y cuál es la ventana temporal de mayor impacto?

4. **Patrones espaciales y clusters de riesgo:** ¿Qué patrones espaciales y clusters territoriales de riesgo pueden identificarse en las comunas de la Región Metropolitana a partir de variables de contaminación y salud?

5. **Predicción de días de alto riesgo:** ¿Es posible predecir días de alto riesgo sanitario utilizando modelos de aprendizaje automático basados en contaminantes atmosféricos y variables temporales?

6. **Umbrales y efectos causales:** ¿Se observan cambios significativos en los eventos de salud respiratoria cuando los niveles de contaminación superan ciertos umbrales normativos, sugiriendo posibles efectos causales o discontinuidades?

## 5. Modelamiento

### 5.1 Modelos estadísticos

Dado que los eventos de salud son variables de conteo:

- Regresión de Poisson
- Regresión binomial negativa
- Modelos aditivos generalizados (GAM)

Estos modelos permiten capturar:
- No linealidades
- Efectos de estacionalidad
- Influencia de variables climáticas

---

### 5.2 Efectos rezagados

Se incorporan variables de tipo:

- \( X_t, X_{t-1}, X_{t-2}, ..., X_{t-k} \)

Esto permite evaluar:
- Impacto diferido de la contaminación
- Ventanas críticas de exposición

---

### 5.3 Técnicas supervisadas

#### 5.3.1 Regresión

- Random Forest Regressor
- Gradient Boosting (XGBoost / LightGBM)

Objetivo:

---

#### 5.3.2 Clasificación

Definición de variable objetivo:

Modelos:
- Logistic Regression
- Random Forest Classifier
- Gradient Boosting

Evaluación:
- Accuracy
- ROC-AUC
- F1-score

---

### 5.4 Técnicas no supervisadas

#### 5.4.1 Clustering

- K-means
- DBSCAN
- Clustering jerárquico

Aplicaciones:
- Agrupar comunas según perfil de contaminación

#### 5.4.2 Reducción de dimensionalidad

Objetivos:

#### 5.4.3 Detección de anomalías

Aplicaciones:

## Atenciones de Urgencia 2024

```python
## Carga de datos
URGENCIA_CACHE = 'data/urgencias/urgencia_2020_2024.parquet'
data_raw = pd.read_parquet(URGENCIA_CACHE, engine='pyarrow')

# Filter to 2024 to match original scope (remove or adjust to use all years)
data_raw['_año'] = pd.to_datetime(data_raw['fecha'], dayfirst=True).dt.year
data_raw = data_raw[data_raw['_año'] == 2024].drop(columns=['_año'])

data_raw.head(4)
```

```python
## Carga de mapa por comuna
# Choropleth animation requires Comunas.zip shapefile — not available in this repo.
# To enable it: place the file at Health-AirQuality_DataMining/01. Mapas/Comunas.zip
# and uncomment the lines below.
#
# gdf_comunas = gpd.read_file('01. Mapas/Comunas.zip')
# gdf_comunas = gdf_comunas[gdf_comunas["Region"] == "Región Metropolitana de Santiago"]
gdf_comunas = None
print("gdf_comunas no cargado — shapefile no disponible. Celda de animación deshabilitada.")
```

```python
## Filtro Causas respiratorias
# El parquet ya está filtrado a RM — no se necesita filtro por CodigoRegion.
# Se filtra por GlosaCausa usando causas del sistema respiratorio.
RESP_PATTERN = r'RESPIRAT|IRA\s*Alta|BRONQUIT|INFLUENZA|NEUMON|ASMA|EPOC'
data = data_raw[data_raw['GlosaCausa'].str.contains(RESP_PATTERN, case=False, na=False)].copy()

print('Cantidad de filas antes del filtro: ', data_raw.shape[0])
print('Cantidad de filas después del filtro: ', data.shape[0])

print('-' * 105)
for col in data.columns:
    print(f'{col}: {data[col].dtype}')

print('\n')
print('-' * 105)
print('-' * 105)

## Cantidad de establecimientos en la RM
print('Cantidad de Establecimientos: ', len(data['IdEstablecimiento'].unique()), '\n')

print('-' * 105)
print('-' * 105)
```

```python
print('-' * 105)
print('-' * 105)

## Cantidad de establecimientos en la RM
# NombreComuna no está disponible en la caché — se agrupa por establecimiento.
print('Establecimientos únicos (NombreComuna no disponible en caché):')
print(
    data
    .groupby('NEstablecimiento')['IdEstablecimiento']
    .nunique()
    .sort_values(ascending=False)
    .reset_index(name='n_est')
)

print('-' * 105)
print('-' * 105)
```

```python
## Series de tiempo Total de causas sistema respiratorio (diario)
ts = (
    data[data['GlosaCausa'] == 'TOTAL CAUSAS SISTEMA RESPIRATORIO']
    .groupby('fecha')['Total']
    .sum()
    .reset_index()
)

ts['fecha'] = pd.to_datetime(ts['fecha'], dayfirst=True)
ts = ts.sort_values('fecha')

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(ts['fecha'], ts['Total'], linewidth=2.5)

ax.set_title(
    'Casos de urgencia por causa respiratoria (Diario)',
    fontsize=16, fontweight='bold', loc='left'
)
ax.set_ylabel('Número de casos', fontsize=11)
ax.set_xlabel('')

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.xticks(rotation=0)
ax.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
## Series de tiempo Total de causas sistema respiratorio (semanal)
ts = (
    data[data['GlosaCausa'] == 'TOTAL CAUSAS SISTEMA RESPIRATORIO']
    .assign(fecha_dt=lambda df: pd.to_datetime(df['fecha'], dayfirst=True))
    .groupby(pd.Grouper(key='fecha_dt', freq='W'))['Total']
    .sum()
    .reset_index()
    .rename(columns={'fecha_dt': 'semana'})
)
ts = ts.sort_values('semana')

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(ts['semana'], ts['Total'], linewidth=2.5)

ax.set_title(
    'Casos de urgencia por causa respiratoria (Semanal)',
    fontsize=16, fontweight='bold', loc='left'
)
ax.set_ylabel('Número de casos', fontsize=11)
ax.set_xlabel('')

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.xticks(rotation=0)
ax.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

## Problema 1: Agregar tasa de habitantes por comuna para generar un resultado correcto y comparable entre comunas

```python
# Series de tiempo por causa respiratoria y rango etario
cols = ['Total', 'Menores_1', 'De_1_a_4', 'De_5_a_14', 'De_15_a_64', 'De_65_y_mas']
ts = (
    data[data['GlosaCausa'] == 'TOTAL CAUSAS SISTEMA RESPIRATORIO']
    .groupby('fecha')[cols]
    .sum()
    .reset_index()
)

ts['fecha'] = pd.to_datetime(ts['fecha'], dayfirst=True)
ts = ts.sort_values('fecha')

fig, ax = plt.subplots(figsize=(12, 6))

for col in cols[1:]:
    ax.plot(
        ts['fecha'], ts[col],
        linewidth=1.5, alpha=0.8, linestyle='--',
        label=col.replace('_', ' ')
    )

ax.set_title(
    'Casos de urgencia por causa respiratoria\nDesagregación por rango etario (diario)',
    fontsize=16, fontweight='bold', loc='left'
)
ax.set_ylabel('Número de casos')
ax.set_xlabel('')

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(frameon=False, ncol=3, fontsize=9)

plt.tight_layout()
plt.show()
```
